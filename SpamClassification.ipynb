# Spam Classifier

```python
#Importing necessary packages for the classification 
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.svm import SVC  
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
#Importing data into a dataframe
spam=pd.read_csv("spambase.data.csv",header=None)
spam_data=spam.drop(57, axis=1)
print(spam_data)
spam_label=spam[57]
```

# Guassian Naive Bayes

```python
#splitting data into train and testing data using kfolds
spam_data=np.array(spam_data)
kfold=KFold(n_splits=3, shuffle=True)
for train_index, test_index in kfold.split(spam_data):
    print("TRAIN:", train_index, "TEST:", test_index)
    training_features,testing_features=spam_data[train_index], spam_data[test_index]
    training_labels,testing_labels=spam_label[train_index], spam_label[test_index]
    #training the training features with the training labels by fitting it into the Multinomial Naive Bayes Algorithm
    gnb = GaussianNB()
    gnb.fit(training_features,training_labels)
    labels_gnb=gnb.predict(testing_features)
    #Metric for Gaussian Naive Bayes Model: Accuracy.
    print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(testing_labels, labels_gnb)*100)
    #confusion Matrix: Guassian Naive Bayes
    conf_matrix=[[]]
    conf_matrix=confusion_matrix(testing_labels, labels_gnb)
    print("False Positive: ",conf_matrix[0][1])
    print("False Negative: ",conf_matrix[1][0])
    print("Confusion Matrix: ")
    print(confusion_matrix(testing_labels, labels_gnb))
    error_rate=100-metrics.accuracy_score(testing_labels, labels_gnb)*100
    print("Error rate for Gaussian Naive Bayes model (in %):", error_rate);
    print("----------------------------------------------------------------")
```

# Multinomial Naive Bayes

```python
kfold=KFold(n_splits=3, shuffle=True)
for train_index, test_index in kfold.split(spam_data):
    print("TRAIN:", train_index, "TEST:", test_index)
    training_features,testing_features=spam_data[train_index], spam_data[test_index]
    training_labels,testing_labels=spam_label[train_index], spam_label[test_index]
    #training the training features with the training labels by fitting it into the Multinomial Naive Bayes Algorithm
    clf = MultinomialNB()
    clf.fit(training_features,training_labels)
    #Metric for Multinomial Naive Bayes Model: Accuracy.
    labels_mnb=clf.predict(testing_features)
    print("Multinomial Naive Bayes model accuracy(in %):", metrics.accuracy_score(testing_labels, labels_mnb)*100)
    #confusion Matrix: Multinomial Naive Bayes
    conf_matrix=confusion_matrix(testing_labels, labels_mnb)
    print("False Positive: ",conf_matrix[0][1])
    print("False Negative: ",conf_matrix[1][0])
    print("Confusion Matrix: ")
    print(confusion_matrix(testing_labels, labels_mnb))  
    error_rate=100-metrics.accuracy_score(testing_labels, labels_mnb)*100
    print("Error rate for Multinomial Naive Bayes model (in %):", error_rate);
    print("----------------------------------------------------------------")
```

# Bernoulli Naive Bayes

```python
kfold=KFold(n_splits=3, shuffle=True)
for train_index, test_index in kfold.split(spam_data):
    print("TRAIN:", train_index, "TEST:", test_index)
    training_features,testing_features=spam_data[train_index], spam_data[test_index]
    training_labels,testing_labels=spam_label[train_index], spam_label[test_index]
    #training the training features with the training labels by fitting it into the Bernoulli Naive Bayes Algorithm
    bnb = BernoulliNB()
    bnb.fit(training_features,training_labels)
    labels_bnb=bnb.predict(testing_features)
    print("Bernoulli Naive Bayes model accuracy(in %):", metrics.accuracy_score(testing_labels, labels_bnb)*100)
    #confusion Matrix: Bernoulli Naive Bayes
    conf_matrix=confusion_matrix(testing_labels, labels_bnb)
    print("False Positive: ",conf_matrix[0][1])
    print("False Negative: ",conf_matrix[1][0])
    print("Confusion Matrix: ")
    print(confusion_matrix(testing_labels, labels_bnb))  
    error_rate=100-metrics.accuracy_score(testing_labels, labels_bnb)*100
    print("Error rate for Bernoulli Naive Bayes model (in %):", error_rate)
    print("----------------------------------------------------------------")
```

# Decision Tree

```python
kfold=KFold(n_splits=3, shuffle=True)
for train_index, test_index in kfold.split(spam_data):
    print("TRAIN:", train_index, "TEST:", test_index)
    training_features,testing_features=spam_data[train_index], spam_data[test_index]
    training_labels,testing_labels=spam_label[train_index], spam_label[test_index]
    #training the training features with the training labels by fitting it into the Decision tree Algorithm
    decisiontree = DecisionTreeClassifier(criterion = "gini", random_state = 100,max_depth=3, min_samples_leaf=5)
    decisiontree.fit(training_features,training_labels)
    #Metric for Decision Tree: Accuracy
    labels_decisiontree=decisiontree.predict(testing_features)
    print("Decision tree model accuracy(in %):", metrics.accuracy_score(testing_labels, labels_decisiontree)*100)
    #confusion Matrix: Decision Tree
    conf_matrix=confusion_matrix(testing_labels, labels_decisiontree)
    print("False Positive: ",conf_matrix[0][1])
    print("False Negative: ",conf_matrix[1][0])
    print("Confusion Matrix: ")
    print(confusion_matrix(testing_labels, labels_decisiontree)) 
    error_rate=100-metrics.accuracy_score(testing_labels, labels_decisiontree)*100
    print("Error rate for Decision Tree model (in %):", error_rate)
    print("----------------------------------------------------------------")
```

# Support Vector Machines (SVM)

```python
kfold=KFold(n_splits=2, shuffle=True)
for train_index, test_index in kfold.split(spam_data):
    print("TRAIN:", train_index, "TEST:", test_index)
    training_features,testing_features=spam_data[train_index], spam_data[test_index]
    training_labels,testing_labels=spam_label[train_index], spam_label[test_index]
    #training the training features with the training labels by fitting it into the Linear SVM
    svclassifier = SVC(kernel='linear')  
    svclassifier.fit(training_features,training_labels)
    labels_svm = svclassifier.predict(testing_features) 
    #Metric for Linear SVM: Accuracy
    print("SVM model accuracy(in %):", metrics.accuracy_score(testing_labels, labels_svm)*100)
    #confusion Matrix: SVM
    conf_matrix=confusion_matrix(testing_labels, labels_svm)
    print("False Positive: ",conf_matrix[0][1])
    print("False Negative: ",conf_matrix[1][0])
    print("Confusion Matrix: ")
    print(confusion_matrix(testing_labels, labels_svm)) 
    error_rate=100-metrics.accuracy_score(testing_labels, labels_svm)*100
    print("Error rate for Decision Tree model (in %):", error_rate)
    print("----------------------------------------------------------------")
```
